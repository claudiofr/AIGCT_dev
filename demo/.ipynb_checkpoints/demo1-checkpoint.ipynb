```python
%matplotlib inline

import context
import pandas as pd
from agct.container import VEBenchmarkContainer

user_test_vep_scores = pd.read_csv("test_vep.csv")

container = VEBenchmarkContainer("..")

analyzer = container.analyzer

metrics = container.analyzer.compute_metrics(
    "cancer", user_test_vep_scores, vep_min_overlap_percent=50,
    variant_vep_retention_percent=1, list_variants=True)

```

```python
container.reporter.write_summary(metrics)
```

```python
container.plotter.plot_results(metrics)
```

```python
container.exporter.export_results(metrics, "./output")
```
